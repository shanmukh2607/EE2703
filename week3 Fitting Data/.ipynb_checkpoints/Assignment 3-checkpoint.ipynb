```python
def g(t,A,B):
    return A*jn(2,t) + B*t
```

```python
figure0 = plt.gcf()
figure0.set_size_inches(10,8)
plt.plot(t,data)
y = g(t,1.05,-0.105)
plt.plot(t,y,label='True Value')
sigma_str.append('True Value')
plt.legend(sigma_str)
plt.xlabel(r'$t$',size = 15)
plt.ylabel(r'$f(t)$',size = 15)
plt.title(r'Plot of True value of f(t) vs t for A = 1.05 and B = -0.105')
plt.grid(True)
plt.show()
```

```python
# Point 7
k =1
a = np.linspace(0,2,21)
b = np.linspace(-0.2,0,21)
mse = np.zeros((21,21))
for i in range(len(a)):
    for j in range(len(b)):
        v = np.array([
            [a[i]],
            [b[j]]
        ])
        g_vec = np.transpose(np.dot(M,v))
        mse[i,j]= np.sum((data[:,k]-g_vec)**2)/101
        
```

```python
ind = np.where(mse == np.amin(mse))
fig ,ax = plt.subplots()
cs = ax.contour(a,b,mse,levels = 16)
ax.clabel(cs,cs.levels[:5], inline =1, fontsize = 9)
ax.plot(a[ind[0][0]],b[ind[1][0]], marker = 'o', color ='r')
ax.annotate('Location of min', xy=(a[ind[0][0]],b[ind[1][0]]))
plt.xlabel(r'$A \longrightarrow$')
plt.ylabel(r'$B \longrightarrow$')
plt.title(r'Q8: Contour Plot of $\epsilon_{ij}$');
```

```python
A_pred = np.zeros(9)
B_pred = np.zeros(9)
for k in range(0,9):
    soln =scipy.linalg.lstsq(M,data[:,k])
    A_pred[k] = soln[0][0]
    B_pred[k] = soln[0][1]
A_err = np.abs(1.05 - A_pred)
B_err = np.abs(0.105 + B_pred)
fig, ax = plt.subplots()
ax.plot(sigma,A_err, 'o--',label ='$Aerr$')
ax.plot(sigma,B_err, 'o--',label ='$Berr$')
plt.xlabel(r'Noise standard deviation')
plt.ylabel(r'Error in estimation')
plt.legend()
plt.title(r'Variation of Error with noise')
plt.grid()
```
