```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(100)
X = np.fft.fft(x)
y = np.fft.ifft(X)
np.c_[x,y]
print(abs(x-y).max())
```

```python
x = np.linspace(0,2*np.pi,128)
y= np.sin(x)
Y = np.fft.fft(y)
plt.plot(np.abs(Y),lw=2)
```

```python
plt.plot(np.angle(Y),lw=2)
plt.grid()
```

```python
x = np.linspace(0,2*np.pi,129)
x= x[:-1]
```

```python
y = np.sin(5*x)
Y = np.fft.fftshift(np.fft.fft(y))/128
w = np.linspace(-64,63,128)
fig,ax = plt.subplots(figsize=(7,6))
plt.plot(w,np.abs(Y))
plt.xlim([-10,10])
plt.grid()
plt.plot(w,np.angle(Y),'ro')
ii = np.where(np.abs(Y)>1e-3)
plt.plot(w[ii],np.angle(Y[ii]),'go')
```

```python
t = np.linspace(-2*np.pi,2*np.pi,257)
t= t[:-1]
y=(1+0.1*np.cos(t))*np.cos(10*t)
Y = np.fft.fftshift(np.fft.fft(y))/256
w = np.linspace(-64,64,257);w=w[:-1]
fig,ax = plt.subplots(figsize=(7,6))
plt.plot(w,np.abs(Y))
plt.xlim([-15,15])
plt.grid()
plt.plot(w,np.angle(Y),'ro')
ii = np.where(np.abs(Y)>1e-3)
plt.plot(w[ii],np.angle(Y[ii]),'go')
```

```python
t = np.linspace(-4*np.pi,4*np.pi,513)
t= t[:-1]
y=(1+0.1*np.cos(t))*np.cos(10*t)
Y = np.fft.fftshift(np.fft.fft(y))/512
w=np.linspace(-64,64,513)
w=w[:-1]
fig,ax = plt.subplots(figsize=(7,6))
plt.plot(w,np.abs(Y))
plt.xlim([-15,15])
plt.grid()
fig,ax = plt.subplots(figsize=(7,6))
plt.plot(w,np.angle(Y),'ro')
ii = np.where(np.abs(Y)>1e-3)
plt.plot(w[ii],np.angle(Y[ii]),'go')
plt.xlim([-15,15])
```

```python
def dft_plotter(tstart,tstop,samples,f,x_lim,title):
    t = np.linspace(tstart,tstop,samples+1)
    t=t[:-1]
    y = f(t)
    Y = np.fft.fftshift(np.fft.fft(y))/samples
    w = np.linspace(-64,64,samples+1)
    w = w[:-1]
    fig ,ax = plt.subplots(figsize=(7,7))
    plt.subplot(2,1,1)
    plt.plot(w,np.abs(Y))
    plt.xlim([-x_lim,x_lim])
    plt.ylabel(r"$|Y|\longrightarrow$")
    plt.title(r"Spectrum of " + title)
    plt.grid()
    plt.subplot(2,1,2)
    plt.plot(w,np.angle(Y),'ro')
    ii = np.where(np.abs(Y)>1e-3)
    plt.plot(w[ii],np.angle(Y[ii]),'go')
    plt.ylabel(r"$\phi$ of $Y\longrightarrow$")
    plt.xlim([-x_lim,x_lim])
    plt.xlabel(r"$\omega \longrightarrow$")
    plt.grid()
```

```python
def gg(t):
    return (1+0.1*np.cos(t))*np.cos(10*t)
```

```python
dft_plotter(-4*np.pi,4*np.pi,512,gg,15,r'$\left(1+0.1\cos\left(t\right)\right)\cos\left(10t\right)$')
```
