```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def exp(x):
    return np.exp(x)

def coscos(x):
    return np.cos(np.cos(x))
```

```python
def periodic_maker(fun):
    def newfunc(x):
        return fun(np.remainder(x,2*np.pi))
    return newfunc

exp_periodic = periodic_maker(exp)
coscos_periodic = periodic_maker(coscos)
```

```python
x = np.linspace(-2*np.pi, 4*np.pi, 1000)

fig, ax = plt.subplots(num=1)
ax.semilogy(x,exp(x),label ='Actual Function')
ax.semilogy(x,exp_periodic(x), label='Expected')
plt.xlabel(r'$x \longrightarrow$', size =15)
plt.ylabel(r'$e^x \longrightarrow$', size =15)
plt.title(r'Semilogy plot of $e^x$ vs $x$ in $[-2\pi,4\pi)$')
plt.grid()

fig, ax = plt.subplots(num=2)
ax.plot(x, coscos(x))
ax.plot(x, coscos_periodic(x))
plt.xlabel(r'$x \longrightarrow$', size =15)
plt.ylabel(r'$cos(cos(x)) \longrightarrow$', size =15)
plt.title(r'Plot of $cos(cos(x))$ vs $x$ in $[-2\pi,4\pi)$')
plt.grid()

```
